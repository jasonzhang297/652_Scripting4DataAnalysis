# Code for data analysis 1

```python
#Subset crimes df of year 2021 only
year2021 = crimes[crimes["Year"] == 2021]
year2021.head()
```

```python
#Unique value counts
year2021["Primary Type"].value_counts()
year2021_count = year2021["Primary Type"].value_counts().to_frame()
year2021_count = year2021_count.reset_index()
year2021_count.columns = ['Crime Type','Count']
year2021_count.head()
```

```python
#Piechart of top5 crime types in year 2021
year2021_count5 = year2021_count.head(5)
plot = year2021_count5.plot.pie(y='Count', figsize=(8, 8))
plt.legend(year2021_count5["Crime Type"])
plt.title("Top5 crime types in Year 2021")
```

## Description of data analysis result 1

My goal here is to show the amount of the top 5 crime types in year 2021 by plotting a piechart. As a result, by looking at the plot, we can clealy see that the top 5 crimes types in year 2021 are theft; battery; criminal damage; assault and deceptive practice. Among all the crimes types: theft has 40746 cases; battery has 40447 cases; criminal damage has 25089 cases; assault has 20332 cases and deceptive practice has 16990 cases.

# Code for data analysis 2

```python
#Subset crimes df of year 2021 an Domestic = true only
DomesticT = crimes[(crimes['Domestic'] == True) & (crimes['Year'] == 2021)]
DomesticT.head()
```

```python
#Return probability of Domestic crimes and been arrested
DomesticT.Arrest.value_counts()[True] / DomesticT.shape[0]
```

```python
#Subset crimes df of year 2021 an Domestic = true only
DomesticF = crimes[(crimes['Domestic'] == False) & (crimes['Year'] == 2021)]
DomesticF.head()
```

```python
#Return probability of Non-Domestic crimes and been arrested
DomesticF.Arrest.value_counts()[True] / DomesticF.shape[0]
```

## Description of data analysis result 2

My goal here is to show the probability of criminals been arrested for both Domestic and Non-Domestic crimes in Year 2021. By looking the results above, we can see that the probability of Domestic crimes and criminals been arrested is about 11.7%; the probability of Non-Domestic crimes and criminals been arrested is about 12.6% which is slightly higher than Domestic crimes. Overall, both condtions have low probability that criminals been arrested which is bad to hear.
